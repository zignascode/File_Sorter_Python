# FileSorter is an script to sort and classify files

```python
#Library imports

import pandas as pd     #For data manipulation and analysis
import os               #For operating system related functions
import hashlib as hl    #For hashing files to check for duplicates
import datetime as dt   #For datetime format
```

```python
#Getting the list of files in the current directory

ls_files = [] #List with only files, no directories

for file in os.listdir():
    if os.path.isfile(file):
        ls_files.append(file)           #If the item is a file, add it to the list
        
num_files=len(ls_files)                 #Total of files in the directory

print(f"{num_files} files found in the currently directory")
```

```python
#Getting the attributes of all of the files

ls_name=[]; ls_ext=[]; ls_hash=[]; ls_size=[]; ls_dateM=[]

for i in range(0,num_files):

    file=ls_files[i]                            #file
    name=os.path.splitext(file)[0]              #Name of the file

    size=os.stat(file).st_size                  #Size of the file

    ext=os.path.splitext(file)[1][1:]           #Type of the extension file
    if ext == "": ext="NULL"                    #If the extension is NULL

    stampM=os.stat(file).st_mtime                #Last change in timestamp format
    dateM=dt.date.fromtimestamp(stampM)      #Last change, in datetime format

    with open(file, "rb") as door:               #Needs to process with "with" to close automaticaly
        hash = hl.sha1(door.read()).hexdigest()  #Calculate the hash object and get with hexdigest type
                                                 #No need update() because the files are simple

    ls_name.append(name)
    ls_ext.append(ext)
    ls_hash.append(hash)
    ls_size.append(size)
    ls_dateM.append(dateM)
```

```python
#This part is for build the dataframe and get the preliminar stats

dc_data = {
    "NAME": ls_name,
    "EXT": ls_ext,
    "HASH": ls_hash,
    "SIZE": ls_size,
    "DATE_M": ls_dateM
}

tb_data=pd.DataFrame(dc_data)
tb_stats=tb_data.groupby("EXT").agg(
    COUNT=("EXT", "count"),
    TOTAL_SIZE=("SIZE", "sum")
).reset_index()

tb_stats.sort_values(by="TOTAL_SIZE", ascending=False)
#tb_data
```
