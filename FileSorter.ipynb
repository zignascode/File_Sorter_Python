# FileSorter is an script to sort and classify files

```python
#Library imports

import pandas as pd #For data manipulation and analysis
import os #For operating system related functions
import hashlib as hl #For hashing files to check for duplicates
```

```python
#Getting the list of files in the current directory

ls_files = [] #List with only files, no directories

for file in os.listdir():
    #If the item is a file, add it to the ls_fix list
    if os.path.isfile(file): ls_files.append(file)

#ls_files
```

```python
#Getting the attributes of all of the files

ls_name=[]
ls_hash=[]
ls_size=[]

for i in range(0,len(ls_files)):
    with open(ls_files[i], "rb") as door: #Needs to process with "with" to close automaticaly
        hash = hl.sha1(door.read()).hexdigest() #Calculate the hash object and get with hexdigest type
        #No need update() because the files are simple

    size=os.stat(ls_files[i]).st_size

    ls_name.append(ls_files[i])
    ls_hash.append(hash)
    ls_size.append(size)
```

```python
#This part is for build the dataframe 

dc_data = {
    "NAME": ls_name,
    "HASH": ls_hash,
    "SIZE": ls_size
}

tb_data=pd.DataFrame(dc_data)
tb_data
```
