# FileSorter is an script to sort and classify files

```python
#Library imports

import pandas as pd     #For data manipulation and analysis
import os               #For operating system related functions
import hashlib as hl    #For hashing files to check for duplicates
import datetime as dt   #For datetime format
```

```python
#Getting the list of files in the current directory

ls_files = [] #List with only files, no directories

for file in os.listdir():
    #If the item is a file, add it to the ls_fix list
    if os.path.isfile(file): ls_files.append(file)

#ls_files
```

```python
#Getting the attributes of all of the files

ls_name=[]
ls_hash=[]
ls_size=[]
ls_dateM=[]
#ls_dateA=[]

for i in range(0,len(ls_files)):

    name=ls_files[i]                             #Name of the file
    size=os.stat(name).st_size                   #Size of the file

    stampM=os.stat(name).st_mtime                #Last change in timestamp format
    dateM=dt.datetime.fromtimestamp(stampM)      #Last change, in datetime format

    #stampA=os.stat(name).st_atime               #Last access in timestamp format
    #dateA=dt.datetime.fromtimestamp(stampA)     #Last access, in datetime format

    with open(name, "rb") as door:               #Needs to process with "with" to close automaticaly
        hash = hl.sha1(door.read()).hexdigest()  #Calculate the hash object and get with hexdigest type
                                                 #No need update() because the files are simple

    ls_name.append(name)
    ls_hash.append(hash)
    ls_size.append(size)
    ls_dateM.append(dateM)
    #ls_dateA.append(dateA)
```

```python
#This part is for build the dataframe 

dc_data = {
    "NAME": ls_name,
    "HASH": ls_hash,
    "SIZE": ls_size,
    "DATE_M": ls_dateM
    #"DATE_A": ls_dateA
}

tb_data=pd.DataFrame(dc_data)
tb_data
```
